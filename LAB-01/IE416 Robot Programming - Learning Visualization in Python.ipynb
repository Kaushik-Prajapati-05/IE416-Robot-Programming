# Bar plots with legend

````python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
categories = ['A', 'B', 'C']
values1 = [10, 20, 15]
values2 = [12, 18, 22]

bar_width = 0.35
index = np.arange(len(categories))

# Plotting the bars
plt.bar(index, values1, bar_width, label='Bar Group 1', color='blue')
plt.bar(index + bar_width, values2, bar_width, label='Bar Group 2', color='red')

# Adding a title and labels
plt.title('Multiple Bar Plots with Legend')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.xticks(index + bar_width / 2, categories)

# Adding the legend
# plt.legend
plt.legend(loc='best', fontsize='small', frameon=True, shadow=True)

# Display the plot
plt.show()
````

# Scatter plots

````python
import matplotlib.pyplot as plt

# Sample data
x1 = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
x2 = [1, 2, 3, 4, 5]
y2 = [1, 4, 6, 8, 10]

# Scatter plot
plt.scatter(x1, y1, label='Group 1', color='blue', marker='o')
plt.scatter(x2, y2, label='Group 2', color='red', marker='x')

# Adding a title and labels
plt.title('Multiple Scatter Plots with Legend')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')

# Adding the legend
plt.legend()

# Display the plot
plt.show()
````

# 3D plots

````python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Generate random data
x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)

# Create a new figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Scatter plot
ax.scatter(x, y, z, c='r', marker='o')

# Labels
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.set_title('3D Scatter Plot')

# Show the plot
plt.show()
````

# Adding subplots in a figure

````python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
# plt.ion()
# Create grid and multivariate normal
X = np.linspace(-5, 5, 100)
Y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(X, Y)
Z = X**2 + Y**2

# Create a new figure
fig = plt.figure()
ax = fig.add_subplot(121, projection='3d')
# Surface plot
ax.plot_surface(X, Y, Z, cmap='viridis')
# Labels
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.set_title('3D Surface Plot')
ax1 = fig.add_subplot(122, projection='3d')
Z = X**2 - Y**2
# Surface plot
ax1.plot_surface(X, Y, Z, cmap='viridis')

# Labels
ax1.set_xlabel('X Label')
ax1.set_ylabel('Y Label')
ax1.set_zlabel('Z Label')
ax1.set_title('3D Surface Plot')



# Show the plot
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Create a grid of points
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Define a function for Z
Z = X**2 + Y**2

# Create a contour plot
plt.figure()
contour = plt.contour(X, Y, Z, levels=20, cmap='viridis')
plt.colorbar(contour)
plt.title('Contour Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
````

# Pie chart

````python
import matplotlib.pyplot as plt

# Data for the pie chart
sizes = [15, 30, 45, 10]  # Sizes of the pie slices
labels = ['Category A', 'Category B', 'Category C', 'Category D']  # Labels for the slices
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen']  # Colors for each slice

# Create a pie chart
plt.figure()
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is circular.
plt.title('Pie Chart')
plt.show()
````

# Box plot

````python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
np.random.seed(10)
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# Create a box plot
plt.figure(figsize=(8, 6))
plt.boxplot(data, vert=True, patch_artist=True,
            labels=['Dataset 1', 'Dataset 2', 'Dataset 3'])

# Customize the plot
plt.title('Box Plot Example')
plt.xlabel('Datasets')
plt.ylabel('Values')
plt.grid(True)

# Show the plot
plt.show()
````

# How to  create  a dual y axis

````python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.arange(0, 10, 0.1)
y1 = np.sin(x)  # First dataset
y2 = np.exp(x)  # Second dataset

# Create a figure and the first axis
fig, ax1 = plt.subplots()

# Plot the first dataset
ax1.plot(x, y1, 'b-', label='sin(x)')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('sin(x)', color='b')
ax1.tick_params(axis='y', labelcolor='b')

# Create a second y-axis
ax2 = ax1.twinx()

# Plot the second dataset
ax2.plot(x, y2, 'r-', label='exp(x)')
ax2.set_ylabel('exp(x)', color='r')
ax2.tick_params(axis='y', labelcolor='r')

# Add a title
plt.title('Dual Y-Axis Example')

# Show the plot
plt.show()
````

# Add a vertical line at a specific position

````python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a plot
plt.plot(x, y)

# Add a vertical line at x = 5
plt.axvline(x=5, color='r', linestyle='--', label='x = 5')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Vertical Line Example')
plt.legend()

# Show the plot
plt.show()
````
